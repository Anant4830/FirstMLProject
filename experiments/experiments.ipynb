Trust scores measure how similar a test instance is to the training data. They can be computed using metrics like distance to nearest neighbors.

Steps:
1. Use a k-nearest neighbors (KNN) model to calculate distances to the nearest neighbors in the training data.
2. Compute the ratio for the trust score.

```python

from sklearn.neighbors import NearestNeighbors
import numpy as np

# Function to compute trust score
def compute_trust_score(instance, X_train, y_train, k=5):
    # Separate training data by class
    same_class = X_train[y_train == y_train[0]]
    different_class = X_train[y_train != y_train[0]]

    # Dynamically adjust k
    k_same = min(k, len(same_class))
    k_diff = min(k, len(different_class))

    # Nearest neighbors for same class
    knn_same = NearestNeighbors(n_neighbors=k_same)
    knn_same.fit(same_class)
    dist_same, _ = knn_same.kneighbors(instance.reshape(1, -1))

    # Nearest neighbors for different class
    knn_diff = NearestNeighbors(n_neighbors=k_diff)
    knn_diff.fit(different_class)
    dist_diff, _ = knn_diff.kneighbors(instance.reshape(1, -1))

    # Compute trust score
    trust_score = np.mean(dist_diff) / np.mean(dist_same)
    return trust_score

# Example usage
index = 0
instance = X_test.iloc[index, :].values
trust_score = compute_trust_score(instance, X_train.values, y_train.values)
print(f"Trust Score for Instance {index}: {trust_score}")
```

```python
import numpy as np
from scipy.spatial.distance import mahalanobis
from sklearn.covariance import EmpiricalCovariance

def compute_mahalanobis_distance(instance, X_train):
    """
    Computes the Mahalanobis distance of a single instance from the training data.
    """
    # Fit an Empirical Covariance Model
    cov_model = EmpiricalCovariance()
    cov_model.fit(X_train)

    # Compute Mahalanobis Distance
    mean = np.mean(X_train, axis=0)
    dist = mahalanobis(instance, mean, cov_model.precision_)
    return dist

# Example usage
index = 0
instance = X_test.iloc[index, :].values
mahalanobis_dist = compute_mahalanobis_distance(instance, X_train.values)
print(f"Mahalanobis Distance for Instance {index}: {mahalanobis_dist}")
```

```python
from pyod.models.knn import KNN
from pyod.models.mcd import MCD

# Train an outlier detection model
def train_pyod_model(X_train):
    """
    Train a PyOD model on the training dataset.
    """
    model = MCD()  # Minimum Covariance Determinant (robust to outliers)
    model.fit(X_train)
    return model

# Compute confidence scores
def compute_pyod_confidence(instance, model):
    """
    Compute the confidence score for a given instance using a trained PyOD model.
    """
    # Reshape instance for model
    instance = instance.reshape(1, -1)
    confidence_score = model.decision_function(instance)  # Higher = More confident
    return confidence_score[0]

# Example usage
pyod_model = train_pyod_model(X_train.values)
index = 0
instance = X_test.iloc[index, :].values
confidence_score = compute_pyod_confidence(instance, pyod_model)
print(f"Confidence Score for Instance {index}: {confidence_score}")
```

```python
def robust_confidence(instance, X_train, pyod_model):
    """
    Combine Mahalanobis distance and PyOD confidence for robust confidence estimation.
    """
    # Mahalanobis Distance
    mahal_dist = compute_mahalanobis_distance(instance, X_train)
    mahal_conf = np.exp(-mahal_dist)  # Convert distance to a confidence score

    # PyOD Confidence
    pyod_conf = compute_pyod_confidence(instance, pyod_model)

    # Combined Confidence (Weighted Average)
    combined_conf = 0.5 * mahal_conf + 0.5 * pyod_conf
    return combined_conf

# Example usage
combined_confidence = robust_confidence(instance, X_train.values, pyod_model)
print(f"Combined Confidence for Instance {index}: {combined_confidence}")
```

```python
import numpy as np
from scipy.spatial.distance import mahalanobis
from sklearn.covariance import EmpiricalCovariance
from pyod.models.mcd import MCD
import shap

def inference_with_explainability_and_trust(
    instance, model, X_train, X_test, pyod_model, conformal_interval, explainer
):
    """
    Perform inference with Explainability, Conformal Prediction, and Trust Scores.

    Parameters:
        - instance: A single instance (1D array) for prediction.
        - model: Trained predictive model (e.g., Linear Regression).
        - X_train: Training data (2D array).
        - X_test: Test data (2D array for conformal prediction).
        - pyod_model: Trained PyOD model for trust score.
        - conformal_interval: Precomputed prediction intervals (e.g., from Quantile Regression).
        - explainer: SHAP explainer object.
    
    Returns:
        - prediction: Model prediction for the instance.
        - explanation: SHAP values for the instance.
        - conformal_bounds: Prediction interval for the instance.
        - mahalanobis_confidence: Confidence score from Mahalanobis distance.
        - pyod_confidence: Confidence score from PyOD.
        - combined_confidence: Combined trust score.
    """

    # Model Prediction
    prediction = model.predict(instance.reshape(1, -1))[0]

    # Explainability: Compute SHAP explanation
    shap_values = explainer.shap_values(instance)
    explanation = shap_values

    # Conformal Prediction: Extract interval for this instance
    instance_index = np.where((X_test == instance).all(axis=1))[0]
    if len(instance_index) == 0:
        raise ValueError("Instance not found in test set for conformal prediction.")
    conformal_bounds = conformal_interval[instance_index[0]]

    # Trust Score: Mahalanobis Distance
    cov_model = EmpiricalCovariance()
    cov_model.fit(X_train)
    mean = np.mean(X_train, axis=0)
    mahal_dist = mahalanobis(instance, mean, cov_model.precision_)
    mahalanobis_confidence = np.exp(-mahal_dist)  # Convert distance to confidence

    # Trust Score: PyOD
    pyod_confidence = pyod_model.decision_function(instance.reshape(1, -1))[0]

    # Combined Trust Score
    combined_confidence = 0.5 * mahalanobis_confidence + 0.5 * pyod_confidence

    return {
        "prediction": prediction,
        "explanation": explanation,
        "conformal_bounds": conformal_bounds,
        "mahalanobis_confidence": mahalanobis_confidence,
        "pyod_confidence": pyod_confidence,
        "combined_confidence": combined_confidence,
    }
```

```python
model_exp.fit(X_train, y_train)

# Train PyOD Model
pyod_model = MCD()
pyod_model.fit(X_train)

# SHAP Explainer
explainer = shap.LinearExplainer(model_exp, X_train)

# Conformal Prediction: Quantile Regression for Prediction Intervals
lower_quantile = np.quantile(y_test - model_exp.predict(X_test), 0.05)
upper_quantile = np.quantile(y_test - model_exp.predict(X_test), 0.95)
conformal_interval = np.array([[pred + lower_quantile, pred + upper_quantile] 
                                for pred in model_exp.predict(X_test)])
#Step 2: Inference with Explainability and Trust Scores
# Test on a single instance
index = 0
instance = X_test.iloc[index, :].values

result = inference_with_explainability_and_trust(
    instance=instance,
    model=model_exp,
    X_train=X_train,
    X_test=X_test,
    pyod_model=pyod_model,
    conformal_interval=conformal_interval,
    explainer=explainer,
)

# Output Results
print(f"Prediction: {result['prediction']}")
print(f"SHAP Explanation: {result['explanation']}")
print(f"Conformal Bounds: {result['conformal_bounds']}")
print(f"Mahalanobis Confidence: {result['mahalanobis_confidence']}")
print(f"PyOD Confidence: {result['pyod_confidence']}")
print(f"Combined Confidence: {result['combined_confidence']}")
```
